```python
# Import des bibliothèques nécessaires
from pyspark.sql import SparkSessionb   
import pymongo

# Création de la session Spark
spark = SparkSession.builder \
    .appName("SparkMongoDBIntegration") \
    .config("spark.mongodb.input.uri", "mongodb://localhost:27017/maBase.maCollection") \
    .config("spark.mongodb.output.uri", "mongodb://localhost:27017/maBase.maCollection") \
    .getOrCreate()

# Connexion à MongoDB via pymongo (optionnel, pour vérification)
client = pymongo.MongoClient("mongodb://localhost:27017/")
print("Bases de données MongoDB disponibles :", client.list_database_names())

# Lecture des données MongoDB avec Spark
df = spark.read.format("mongo").load()

# Affichage des données
df.show()

# Arrêt de la session Spark
spark.stop()

```
